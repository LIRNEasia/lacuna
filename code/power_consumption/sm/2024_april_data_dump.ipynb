```python
root_folder = '/lirneasia/data/lacuna/processed/sm_april_2024_dump'

all_summaries = pd.DataFrame()

folder_path = os.path.join(root_folder)
    
if os.path.exists(folder_path):
    files = os.listdir(folder_path)
        
    for file_name in files:
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
                
            data = pd.read_csv(file_path, low_memory=False)

            data['date'] = pd.to_datetime(data['date'])
            data['year_month'] = data['date'].dt.to_period('M')

            summary_table = data.groupby(['customerref', 'year_month']).size().reset_index(name='counts')
            summary = summary_table.pivot_table(index='customerref', columns='year_month', values='counts').fillna(0)
            summary = summary.astype(int)

            all_summaries = pd.concat([all_summaries, summary], ignore_index=False)
            #df.to_csv(f'/lirneasia/data/lacuna/processed/sm_april_2024_dump/{folder_name}{file_name}.csv', index=False)
        
else:
    print('file path doesnt exist')
```
